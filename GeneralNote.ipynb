```python
import qiskit
```

```python
qr = qiskit.QuantumRegister(2, name = '\psi')
cir = qiskit.QuantumCircuit(qr)
cir.h(0)
cir.cx(0,1)
cir.draw(output='mpl')
```

```python
qr = qiskit.QuantumRegister(6, name='\psi')
cr = qiskit.ClassicalRegister(6, name='c')
circuit = qiskit.QuantumCircuit(qr,cr)
circuit.h(range(6))
circuit.draw()
```
