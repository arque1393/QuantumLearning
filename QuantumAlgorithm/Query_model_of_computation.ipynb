## Query model of Computation 

Input of Query model isd representated by  $\large{f:\Sigma^{n} \rightarrow \Sigma^{m}}$  where $\Sigma \in \{  0,1 \} $ <br>
**Cost of Query Model** : Number of Query is denoted the Cost of Query Computation model


### Example of Query Problems

#### Or Problem 
|||
|--------------|-------------------------------|
| **Input** | $f:\Sigma^n \rightarrow \Sigma$ |
| **Output** | $1$  If there exist a string $x \in \Sigma^n$ for which $f(x)=1$<br>  $0$ If there exist no such String|

#### Parity Problem 
|||
|--------------|-------------------------------|
| **Input** | $f:\Sigma^n \rightarrow \Sigma$ |
| **Output** |$0$ If $f(x)=1$  for an even length of strings $x \in \Sigma^n$ <br>$1$  If $f(x)=1$  for an odd length of strings $x \in \Sigma^n$|

#### Minimum Problem 
|||
|--------------|-------------------------------|
| **Input** | $f:\Sigma^n \rightarrow \Sigma^m$ |
| **Output** | The string $y \in \{f(x):x\in\Sigma^n\}$ that comes <br>first in the lexicographic ordering of $\Sigma^m$|

#### Unique Search Problem 
|||
|--------------|-------------------------------|
| **Input** | $f:\Sigma^n \rightarrow \Sigma$ |
| **Promise** | There is exacly one $z\in\Sigma^n$ for which $f(z)=1$, <br> with $f(x)=0\;\; \forall x \neq z$| 
| **Output** | The $z$ string|
